```python
# Laplacian of Gaussian (LoG)
def laplacian_of_gaussian(img: torch.Tensor, sigma: float = 1.0) -> torch.Tensor:
    kernel_size = int(2 * np.ceil(3 * sigma) + 1)
    x = torch.arange(kernel_size) - kernel_size // 2
    y = torch.arange(kernel_size) - kernel_size // 2
    X, Y = torch.meshgrid(x, y, indexing="ij")

    gaussian = torch.exp(-(X**2 + Y**2) / (2 * sigma**2))
    gaussian /= gaussian.sum()

    laplacian_kernel = (X**2 + Y**2 - 2 * sigma**2) * gaussian / (sigma**4)

    laplacian_kernel = laplacian_kernel.unsqueeze(0).unsqueeze(0)
    img = img.unsqueeze(0).unsqueeze(0)

    output = conv2d(img, laplacian_kernel, padding=kernel_size // 2, stride=1)

    return output.squeeze(0).squeeze(0)


sigma = 1.0
img_laplace = laplacian_of_gaussian(img, sigma)
plt.figure(figsize=(16, 8))
plt.subplot(1, 2, 1)
plt.title("Original Image")
plt.imshow(img, cmap="gray")
plt.axis("off")
plt.subplot(1, 2, 2)
plt.title("Laplacian of Gaussian Image")
plt.imshow(img_laplace, cmap="gray")
plt.axis("off")
plt.show()
```
